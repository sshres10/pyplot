## Question 1: Frequency Bar Plot

1. Given the text below, create a bar plot (or a horizontal bar plot) that shows the frequency distribution of characters in the string. 
A dictionary has been created for you containing the characters and the frequencies of the characters.
Make sure your bar plot has the bars labelled correctly (with the character it represents the frequency of).


```python
from collections import Counter
import matplotlib.pyplot as plt

# Text for character frequency analysis
TEXT = "Continuing into the next season, a special gold version of the Premier League trophy was commissioned to commemorate Arsenal winning the title without a single defeat".lower()

# Calculate the frequency of each character
letter_freq = dict(Counter(TEXT))

# Plotting the bar chart
plt.figure(figsize=(12, 6))
plt.bar(letter_freq.keys(), letter_freq.values())
plt.title("Character Frequency Distribution")
plt.xlabel("Characters")
plt.ylabel("Frequency")
plt.show()

```

## Question 2: Generate and Plot Random Numbers

2. Generate a sequence of at least 5 random numbers (integers are fine), and visualize the data using both `scatter` and `plot` on different axes.  You may use the index of the value as its x value in the plot.

```python
import random
import matplotlib.pyplot as plt

# Generate a sequence of 5 random numbers
random_numbers = [random.randint(0, 100) for _ in range(5)]

# Line plot
plt.figure(figsize=(8, 4))
plt.plot(random_numbers, marker='o', linestyle='-', color='blue')
plt.title("Random Numbers - Line Plot")
plt.xlabel("Index")
plt.ylabel("Random Number")
plt.show()

# Scatter plot
plt.figure(figsize=(8, 4))
plt.scatter(range(len(random_numbers)), random_numbers, color='red')
plt.title("Random Numbers - Scatter Plot")
plt.xlabel("Index")
plt.ylabel("Random Number")
plt.show()
```

## Question 3: Modify Style and Color

3. Do the same question as above, but change the syle and color of both the scatter and the plot.

```python
# Styled line plot
plt.figure(figsize=(8, 4))
plt.plot(random_numbers, marker='^', linestyle='--', color='green')
plt.title("Styled Random Numbers - Line Plot")
plt.xlabel("Index")
plt.ylabel("Random Number")
plt.show()

# Styled scatter plot
plt.figure(figsize=(8, 4))
plt.scatter(range(len(random_numbers)), random_numbers, color='purple', marker='x')
plt.title("Styled Random Numbers - Scatter Plot")
plt.xlabel("Index")
plt.ylabel("Random Number")
plt.show()
```
